```python
def get_context_vec(sentence, target_word, model, window_size=2):
    words = simple_preprocess(' '.join(sentence))
    target_index = words.index(target_word)

    context_words = words[max(0, target_index - window_size):target_index] + \
                    words[target_index + 1:target_index + 1 + window_size]

    # Use 'model.wv.key_to_index' to check if the word is in the vocabulary
    context_vectors = [model.wv[word] for word in context_words if word in model.wv.key_to_index]

    if context_vectors:
        context_vector = np.mean(context_vectors, axis=0)
    else:
        context_vector = np.zeros(model.vector_size)

    return context_vector

def get_sense_vec(glosses, model):
    gloss_vectors = []

    for gloss in glosses:
        gloss_words = simple_preprocess(gloss)
        vectors = [model.wv[word] for word in gloss_words if word in model.wv.key_to_index]
        if vectors:
            gloss_vectors.append(np.mean(vectors, axis=0))

    if gloss_vectors:
        sense_vector = np.mean(gloss_vectors, axis=0)
    else:
        sense_vector = np.zeros(model.vector_size)

    return sense_vector
```

```python
from gensim.models import FastText
from gensim.utils import simple_preprocess
from scipy.spatial.distance import cosine
import numpy as np

# Assuming the previous FastText model and other functions are already defined

# Step 3: Sense Disambiguation with Actual Sense Output
def disambiguate_sense(sentence, target_word, glosses, model):
    context_vector = get_context_vectors(sentence, target_word, model)

    best_similarity = -1  # Initialize with a very low similarity score
    best_gloss = None     # Initialize with no gloss

    for gloss in glosses:
        sense_vector = get_sense_vector([gloss], model)
        similarity = compute_similarity(context_vector, sense_vector)
        print(f"Similarity between context and sense vector: {similarity}")
        # Check if this gloss has a higher similarity
        if similarity > best_similarity:
            best_similarity = similarity
            best_gloss = gloss

    return best_gloss, best_similarity
```

```python
best_gloss, best_similarity = disambiguate_sense(sent, target_word, glosses, model)
print(f"Best sense: {best_gloss} with similarity: {best_similarity}")
```
